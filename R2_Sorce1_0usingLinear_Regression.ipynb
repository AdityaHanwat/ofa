```python
#Load the dataset into a pandas dataframe
df = pd.read_csv('/content/AI_Syantheticdata - Sheet1 (1).csv')

#Handle missing values
imputer = SimpleImputer(strategy='mean')
df[['Yield(in quantial)', 'Price of Crop(INR/Quintal)']] = imputer.fit_transform(df[['Yield(in quantial)', 'Price of Crop(INR/Quintal)']])

#Convert categorical variables to numerical values
le = LabelEncoder()
df['Crop'] = le.fit_transform(df['Crop'])
df['Soil'] = le.fit_transform(df['Soil'])
df['Farmer'] = le.fit_transform(df['Farmer'])
df['State/Region'] = le.fit_transform(df['State/Region'])
df['Season'] = le.fit_transform(df['Season'])
df['Variety of crop'] = le.fit_transform(df['Variety of crop'])
df['Soil Condition'] = le.fit_transform(df['Soil Condition'])

#Scale the numerical features
scaler = StandardScaler()
df[['Yield(in quantial)', 'Price of Crop(INR/Quintal)']] = scaler.fit_transform(df[['Yield(in quantial)', 'Price of Crop(INR/Quintal)']])

#Split the preprocessed data into training and testing sets
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(df)
```

```python
#Train the model on the training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Make predictions on the testing data
y_pred = regressor.predict(X_test)

#Calculate the performance metrics
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("R2 score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
